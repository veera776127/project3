# Data Columns

- **Fruit:** This column lists the names of the fruits. It includes both common names and specific types or preparations, such as "Apples, applesauce" or "Apples, ready-to-drink".
  
- **Form:** It indicates the form in which the fruit is sold or consumed, such as 'Fresh', 'Canned', or 'Juice'.
  
- **RetailPrice:** This column shows the retail price of each fruit in its respective form.
  
- **RetailPriceUnit:** It denotes the unit of measurement for the retail price, for example, 'per pound' or 'per pint'.
  
- **Yield:** This column provides the yield of the fruit, which can be a measure of the usable product after preparation or processing.
  
- **CupEquivalentSize:** It indicates the size of a standard cup equivalent for the fruit, which can be useful for understanding portion sizes.
  
- **CupEquivalentUnit:** This denotes the unit of measurement for the cup equivalent size, such as 'pounds' or 'fluid ounces'.
  
- **CupEquivalentPrice:** This column calculates the price per cup equivalent, offering a standardized measure to compare the cost across different fruits and forms.

### Figure: Mean Retail Price vs. Yield by Fruit

```python
# Research Question : Price Distribution of Top 5 Fruits
top_5_fruits = fruit_prices_df['Fruit'].value_counts().head(5).index
top_5_data = fruit_prices_df[fruit_prices_df['Fruit'].isin(top_5_fruits)]
plt.figure(figsize=(10, 8))
sns.boxplot(x='Fruit', y='RetailPrice', data=top_5_data)
plt.title('Price Distribution Across Top 5 Fruits')
plt.show()
```

### Figure: Price Distribution Across Top 5 Fruits

The boxplot illustrates the price distribution for the top five fruits in the dataset. It allows for comparison of median prices and the spread of prices for each fruit, highlighting variability and potential outliers.


```python
# Research Question : Correlation Heatmap for Different Fruit Forms
for form in fruit_prices_df['Form'].unique():
    plt.figure(figsize=(8, 6))
    form_data = fruit_prices_df[fruit_prices_df['Form'] == form]
    sns.heatmap(form_data[['RetailPrice', 'Yield', 'CupEquivalentSize']].corr(), annot=True, cmap='coolwarm')
    plt.title(f'Correlation Matrix for {form}')
    plt.show()
```

### Figure: Correlation Matrix for Dried

This correlation matrix visualizes the relationships between the retail price, yield, and cup equivalent size specifically for dried fruits. The color gradient indicates the strength and direction of correlations, with warmer colors showing positive and cooler colors showing negative correlations.

## Conclusion on Fruit Price Analysis

In conclusion, our comprehensive analysis of the fruit dataset has revealed several key insights:

- **Price Variability**: There is a significant variability in the retail prices of fruits, as shown by the boxplots. Dried fruits, in particular, have higher retail prices on average compared to other forms, suggesting a premium for the process of drying or the convenience of storage and longevity.

- **Yield Consistency**: The yield for most fruits is consistently high, especially for fresh forms, indicating a lower level of waste and higher consumable proportions. The box plot of yield values confirms this trend across the dataset.

- **Size and Price Correlation**: There is a noticeable negative correlation between cup equivalent size and retail price, which may indicate that smaller quantity purchases or servings are often more expensive. This could be due to packaging and handling costs being higher per unit for smaller sizes.

- **Historical Data Implications**: The correlation matrix provided insights specific to dried fruits, highlighting the different factors influencing their pricing structure compared to other fruit forms.

- **Pricing Strategy Indicators**: The scatter plots and mean price comparisons across different fruits and forms suggest strategic pricing is likely based on yield, form, and perhaps the convenience factor associated with certain fruit types.

These findings can aid stakeholders in the fruit industry, from suppliers to retailers, in making informed decisions about pricing strategies, marketing, and inventory management. Additionally, consumers can use this information to understand price points and make more economical choices when purchasing fruits.

It is important to consider seasonal variations, geographical factors, and market trends for a more granular analysis and to predict future pricing dynamics accurately.
